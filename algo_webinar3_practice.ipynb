1*. Измените функцию calc_logloss так, чтобы нули по возможности не попадали в np.log.

```python
def calc_logloss(y, y_pred):
    y_pred = np.array([(a if ((a!=0) or (a!=1)) else np.abs(a - 1e-8)) for a in y_pred]) #Хотя вообще так делать не следует -_-
    err = - np.mean(y * np.log(y_pred) + (1.0 - y) * np.log(1.0 - y_pred))
    return err
```

```python
def sigmoid(z):
    res = 1 / (1 + np.exp(-z))
    return res
```

2. Подберите аргументы функции eval_model для логистической регрессии таким образом, чтобы log loss был минимальным.

```python
def eval_model(X, y, iterations, alpha=1e-4, metric: callable = None):
    np.random.seed(42)
    W = np.random.randn(X.shape[0])
    n = X.shape[1]
    for i in range(1, iterations+1):
        z = np.dot(W, X)
        y_pred = sigmoid(z)
        err = calc_logloss(y, y_pred)
        W -= alpha * (1/n * np.dot((y_pred - y), X.T))
        if i % (iterations / 10) == 0:
            print(i, W, err)
    return W
```

```python
W = eval_model(X, y, iterations=30000, alpha=1e-1)
```

4. Создайте функцию calc_pred, возвращающую предсказанный класс (на вход подаются W, который уже посчитан функцией eval_model и X, на выходе - массив y_pred).

```python
def calc_pred(W,X):
    thresh = 0.5
    z = np.dot(X.T, W)
    y_pred = sigmoid(z) - thresh
    class_pred = np.array([(1 if (a>=0) else 0) for a in y_pred])
    return class_pred
```

```python
y_pred = calc_pred(W,X)
```

7*. Создайте функции eval_model_l1 и eval_model_l2 с применением L1 и L2 регуляризаций соответственно.

```python
def eval_model_l1(X, y, iterations, alpha=1e-4, lambda_=1e-6, metric: callable = None):
    np.random.seed(42)
    W = np.random.randn(X.shape[0])
    n = X.shape[1]
    for i in range(1, iterations+1):
        z = np.dot(W, X)
        y_pred = sigmoid(z)
        err = calc_logloss(y, y_pred) + lambda_ * np.sum(np.abs(W))
        W -= alpha * (1/n * np.dot((y_pred - y), X.T))
        if i % (iterations / 10) == 0:
            print(i, W, err)
    return W
```

```python
W = eval_model_l1(X, y, iterations=30000, alpha=1e-1, lambda_=1e-3)
```

```python
def eval_model_l2(X, y, iterations, alpha=1e-4, lambda_=1e-6, metric: callable = None):
    np.random.seed(42)
    W = np.random.randn(X.shape[0])
    n = X.shape[1]
    for i in range(1, iterations+1):
        z = np.dot(W, X)
        y_pred = sigmoid(z)
        err = calc_logloss(y, y_pred)
        W -= alpha * (1/n * np.dot((y_pred - y), X.T)) + lambda_ * np.sum(W**2)
        if i % (iterations / 10) == 0:
            print(i, W, err)
    return W
```

```python
W = eval_model_l2(X, y, iterations=30000, alpha=1e-1, lambda_=1e-4)
```
